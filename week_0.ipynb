```python
class problem1 :

    def __init__(self) :
        self.left = ['사람',  '염소', '늑대','배추']
        self.right = []
        self.ship = []
        # print(self.left, self.right, self.ship)

        self.result = -1
        self.left_history = []
        self.right_history = []
        self.ship_history = []
        self.direction_history = []
        self.ship_move_count = 0
        self.direction = 'right'
        self.ship_member = ""


    def move(self) :
        self.left_history.append(self.left)
        self.right_history.append(self.right)
        self.ship_history.append(self.ship)
        self.direction_history.append(self.direction)        
        print('left : ', self.left, 'ship : ', self.ship, 'right : ', self.right, 'direction : ', self.direction)
        self.ship_move_count += 1
        if self.direction == 'right' :
            self.direction = 'left'
            self.right += self.ship
            self.ship.clear()

        else : # direction == left
            self.direction = 'right'
            self.left += self.ship
            self.ship.clear()

    def constraints(self) :
        self.left.sort()
        self.right.sort()
        self.ship.sort()
        if self.left == ['배추', '염소'] or self.right == ['배추', '염소'] :
            return False

        elif self.left == ['늑대', '염소'] or self.right == ['늑대', '염소'] :
            return False

        elif len(self.ship) > 3 or '사람' not in self.ship:
            return False

        else :
            return True



    def problem1(self) :
        count = random.randint(1, 5)
        
        while(True) :
            

            
            if len(self.left) == 0 and len(self.right) == 4 : # 종료 조건
                result = 1
                print("ship moved times", self.ship_move_count)
                # print(self.left_history, self.ship_history, self.right_history)
                return self.left, self.ship, self.right

            else :

                if self.direction == 'right' :
                    self.left.remove('사람')
                    self.ship += ['사람']

                    if random.random() > 0.5 and len(self.left) > 0: # greedy

                        self.ship_member = random.sample(self.left, 1) # sampling or bfs or dfs
                        self.ship += self.ship_member
                        self.left.remove(self.ship_member.pop())

                    if self.constraints() :
                        
                        count = -1
                        self.move()


                    else : 
                        count += 1
                        self.left += self.ship
                        self.ship.clear()
                        
                        continue
                    

                elif self.direction == 'left' :
                    self.right.remove('사람')
                    self.ship += ['사람']

                    if random.random() > 0.5 and len(self.right) > 0:
                        self.ship_member = random.sample(self.right, 1)
                        self.ship += self.ship_member
                        self.right.remove(self.ship_member.pop())

                    if self.constraints() :
                        count = -1
                        self.move()


                    else : 
                        count += 1
                        self.right += self.ship
                        self.ship.clear()
                        continue                                        

```

#### 하노이 문제 ####
> 제약조건 1. 한 타임스탬프 당 한 번의 원판 움직임만 가능  
> 제약조건 2. 위의 원반은 반드시 밑 원반보다 크기가 작아야 함.  
> 종료조건. 모든 원반을 다른 위치의 기둥으로 옮긴 경우

```python
from collections import deque
n = 10

a = deque([i + 1 for i in range(n)])
b = deque([])
c = deque([])

def hanoi(n, a, b, c) :
    
    if n == 1 :
        c.appendleft(a.popleft())
        print('a : ', a, 'b : ', b, 'c : ', c)
        return
    
    else :
        hanoi(n-1, a, c, b)
        print('a : ', a, 'b : ', b, 'c : ', c)
        c.appendleft(a.popleft())
        hanoi(n-1, b, a, c)
        print('a : ', a, 'b : ', b, 'c : ', c)
        


hanoi(n, a, b, c)
print(a, b, c)
```
